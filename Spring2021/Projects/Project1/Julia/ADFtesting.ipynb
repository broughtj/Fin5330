```julia
using StatsKit.HypothesisTests
using StatsPlots
```

```julia
?ADFTest
```

```julia
## Comment: here we are simulating a prior a random walk with drift series
##          (we know before hand that it is a unit-root nonstationary process)
##          (we should then fail to reject the Null of the ADF test)

N = 500
μ = 0.12
y = zeros(N)
y[1] = randn()

for t = 2:N
    y[t] = μ + y[t-1] + randn()
end
```

```julia
plot(y)
```

<br>

* Recall that under the Null hypothesis the time series we are testing is assumed to be subject to unit-root nonstationarity
* Therefore, for a price series or log-price series we probably expect fail to reject 
* For a first-differenced series or a log-first-differenced series we probably expect to reject

<br>

```julia
ADFTest(y,:none,10)
```

```julia
ADFTest(diff(y), :none, 10)
```
