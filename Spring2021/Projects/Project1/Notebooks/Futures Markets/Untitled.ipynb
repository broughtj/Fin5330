```julia
using Flux
using Random
```

```julia
Random.seed!(1234)

β = [10., 20., 30.] # true parameters

predict(x) = (x*β)

function loss(x, y)
    ŷ = predict(x)
    sum((ŷ .- y).^2)
end

x = hcat(rand(500,2), ones(500,1)) # create data

y = x*β .+ 10*rand() # population equation
```

```julia
θ = Flux.Params([β])
gs = Flux.gradient( () -> loss(x, y), θ )
opt = Flux.Optimise.Descent()
```

```julia
Flux.@epochs 5000 Flux.train!(loss, θ, [(x, y)], opt);
```
