# __The Basics of Monte Carlo Simulation__

<br>

```julia
using StatsKit
using StatsPlots
```

<br>

This Monte Carlo simulation is meant to demonstrate by simulation the principles behind the Central Limit Theorem.

* __NB:__ this is typically done with the calculus

* We want to drive home the sampling theory concepts!!

<br>

__Step 1:__ Start with a known population

<br>

```julia
θ = 2.
```

```julia
d = Exponential(θ)
```

```julia
plot(d, fill=(0, .5,:orange))
```

<br>

__Step 2:__ Draw a sample from this population distribution.

<br>

```julia
N = 1000
data = rand(d, N);
```

```julia
histogram(data)
```

```julia
(mean(data), var(data), std(data))
```

<br>

Let's demonstrate by simulation the convergence argument of the Central Limit Theorem. 

<br>

```julia
M = 1_000_000
xbar = zeros(M)

for i in 1:M
    thesample = rand(d, N)
    xbar[i] = mean(thesample)
end
```

```julia
histogram(xbar)
```

_We just did a computational "proof" of the Central Limit Theorem._

```julia
xbar[500]
```

```julia
(minimum(xbar), maximum(xbar))
```

```julia
(mean(xbar), std(xbar))
```

__Q:__ How would we turn this into a Standard Normal distribution?

```julia
zbar = (xbar .- mean(xbar)) / std(xbar);
```

```julia
histogram(zbar)
```

```julia
(mean(zbar), std(zbar))
```

```julia
xbar[500]
```

## Part II - Let's Redo This for a Poisson Distribution

```julia
λ = 5.
dist = Poisson(λ)
```

```julia
plot(dist, fill=(0, .5,:orange))
```

```julia
y = rand(dist, 10000);
```

```julia
histogram(y)
```

```julia
N = 100
M = 500_000
ybar = zeros(M)

for i in 1:M
    draws = rand(dist, N)
    ybar[i] = mean(draws)
end
```

```julia
histogram(ybar)
```
