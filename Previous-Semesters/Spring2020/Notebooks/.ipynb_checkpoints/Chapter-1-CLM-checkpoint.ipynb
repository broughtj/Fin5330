### Prices, Returns and Compounding

__Definitions:__

<br>


___Simple Net Return:___


$$
R_{t} = \frac{P_{t} - P_{t-1}}{P_{t-1}}
$$


<br>


___Simple Gross Return:___


$$
1 + R_{t}
$$

<br>

Gross return from $t-k$ to $t$ (draw timeline on whiteboard):

$$
\begin{align}
1 + R_{t}(k) &= (1 + R_{t}) (1 + R_{t-1}) \cdots (1 + R_{t-k}) \\
             &= \frac{P_{t}}{P_{t-1}} \frac{P_{t-1}}{P_{t-2}} \frac{P_{t-2}}{P_{t-3}} \cdots \frac{P_{t-k+1}}{P_{t-k}} \\
             &= \frac{P_{t}}{P_{t-k}}
\end{align}
$$

<br>

These multiperiod returns are called __compound returns__

* returns are scale-free

* __Not__ unitless (always specific to a time interval)

* should call a rate of return (a flow variable)

<br>

Multiyear returns are often annualized to be comparable

* We will assume this unless stated otherwise

* e.g. $20\%$ means _"annual return of $20\%$"_

<br>

$$
Annualized[R_{t}(k)] = \left[ \Pi_{j=0}^{k-1} (1 + R_{t-j}) \right]^{\frac{1}{k}} - 1
$$

<br>

First order approximation: 

$$
Ann[R_{t}(k)] \approx \frac{1}{k} \sum_{j=0}^{k-1} R_{t-j}
$$

__Continuous Compounding:__

<br>

$$
\begin{align}
r_{t} &= \ln{(1 + R_{t})} = \ln{(\frac{P_{t}}{P_{t-1}})} \\
      &= \ln{(P_{t})} - \ln{(P_{t-1})} \\
      &= p_{t} - p_{t-1} \\
\end{align}
$$

where $p_{t} = \ln{(P_{t})}$

<br>

__NB:__ to distinguish between $R_{t}$ and $r_{t}$ we refer to $R_{t}$ as a simple return

<br>

One advantage of continuously compounded returns becomes clear in multiperiod returns

$$
\begin{align}
r_{t}(k) &= \ln{(1 + R_{t}(k)} = \ln{(1 + R_{t-1})} \ln{(1 + R_{t-2})} \cdots \ln{(1 + R_{t-k+1})} \\
         &= r_{t} + r_{t-1} + r_{t-2} + \cdots + r_{t-k+1} \\
\end{align}
$$

<br>

But a disadvantage comes in calculating portfolio returns

$$
R_{p,t} = \sum_{i=1}^{N} \omega_{i,p} R_{i,t}
$$

whereas

$$
r_{p,t} \ne \sum_{i=1}^{N} \omega_{i,p} r_{i,t}
$$

<br>

__NB:__ Instead you would have to calculate a portfolio _Net Asset Value_ (NAV) in currency units and then take logs


### Dividend Payments

<br>

$$
R_{t} = \frac{P_{t} - D_{t}}{P_{t-1}} - 1
$$

<br>

$$
r_{t} = \ln{(P_{t} + D_{t})} - \ln{(P_{t-1})}
$$


### Excess Returns

<br>

$$
z_{i,t} = R_{i,t} - R_{0,t}
$$

where $R_{0,t}$ is a return on a reference asset (such as a risk-free rate).

<br>

$$
z_{i,t} = r_{i,t} - r_{0,t}
$$

__NB:__ Can think of excess return as the payoff on an arbitrage portfolio that goes long asset $i$ and short asset $0$.
